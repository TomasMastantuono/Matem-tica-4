```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función periódica
def f(x):
    if x >= 0 and x < np.pi/2:
        return 1
    elif x >= np.pi/2 and x < np.pi:
        return 0

# Parámetros
a = 0  # Inicio del intervalo
b = np.pi   # Fin del intervalo
T = b - a  # Período de la función
N = 200  # Número de términos en la serie de Fourier
L = (b - a) / 2

# Función para calcular los coeficientes a_n
def a_n(n, a, b, f):
    result = 0
    return result

# Función para calcular los coeficientes b_n
def b_n(n, a, b, f):
    result = (1 - (-1)**n)/(n*np.pi)
    return result

# Calcular los coeficientes a_0, a_n y b_n
a0 = 1/2
a_coeffs = [a_n(n, a, b, f) for n in range(1, N + 1)]
b_coeffs = [b_n(n, a, b, f) for n in range(1, N + 1)]

# Función para calcular la serie de Fourier
def fourier_series(x, a0, a_coeffs, b_coeffs, a, b):
    result = a0
    for n in range(1, len(a_coeffs) + 1):
        result += (a_coeffs[n-1] * np.cos(n * np.pi * x / L)) + (b_coeffs[n-1] * np.sin(n * np.pi * x / L))
    return result

# Generar puntos para graficar la función original y la serie de Fourier
x_vals = np.linspace(a, b, 1000)
f_vals = [f(x) for x in x_vals]
fourier_vals = [fourier_series(x, a0, a_coeffs, b_coeffs, a, b) for x in x_vals]

# Graficar la función original y la serie de Fourier
plt.plot(x_vals, f_vals, label='Función Original')
plt.plot(x_vals, fourier_vals, label='Serie de Fourier', linestyle='--')
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Aproximación de la Serie de Fourier')
plt.show()
print(a0)
```
